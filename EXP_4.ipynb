```python
# Problem 1 :
"""A student's exam result is stored un a tuple containing three fixed fields : Subject Name, Marks obtained, and grade"""
# Solution :

Student1 = ("Maths",80,"A")
Subject_Student1 = Student1[0]
Marks_Student1 = Student1[1]
Grade_Student1 = Student1[2]

Student2 = ("Physics",75,"B")
Subject_Student2 = Student2[0]
Marks_Student2 = Student2[1]
Grade_Student2 = Student2[2]

Student3 = ("Chemistry",90,"A")
Subject_Student3 = Student3[0]
Marks_Student3 = Student3[1]
Grade_Student3 = Student3[2]

print("Student 1 :")
print("Subject : ",Subject_Student1)
print("Marks : ",Marks_Student1)
print("Grade : ",Grade_Student1)

if Marks_Student1 >= 75:
    print("Student 1 got Distinction")
else:
    print("Student didn't get Distinction")

print("Student 2 :")
print("Subject : ",Subject_Student2)
print("Marks : ",Marks_Student2)
print("Grade : ",Grade_Student2)

if Marks_Student2 >= 75:
    print("Student 2 got Distinction")
else:
    print("Student didn't get Distinction")

print("Student 3 :")
print("Subject : ",Subject_Student3)
print("Marks : ",Marks_Student3)
print("Grade : ",Grade_Student3)

if Marks_Student3 >= 75:
    print("Student 3 got Distinction")
else:
    print("Student didn't get Distinction")
```

```python
# Problem 2 :
"""An organization records an employee's daily attendance for a week using tuple, where "P" represent present and "A" represent absent"""
# Solution :

Employee1 = ("P","P","P","A","P","P","P")
Employee2 = ("P","P","P","P","P","P","P")
Employee3 = ("A","A","A","P","P","P","P")
Employee4 = ("A","A","A","A","A","A","A")

if "A" in Employee1:
    print("\nEmployee 1 :")
    print("Employee1 is present only on : ",Employee1.count("P")," days")
    print("Employee1 is absent only on : ",Employee1.count("A")," days")
else:
  print("Employee present whole the week")

if "A" in Employee2:
    print("\nEmployee 2 :")
    print("Employee1 is present only on : ",Employee2.count("P")," days")
    print("Employee1 is absent only on : ",Employee2.count("A")," days")
else:
  print("Employee present whole the week")

if "A" in Employee3:
    print("\nEmployee 3 :")
    print("Employee1 is present only on : ",Employee3.count("P")," days")
    print("Employee1 is absent only on : ",Employee3.count("A")," days")
else:
  print("Employee present whole the week")

if "A" in Employee4:
    print("\nEmployee 1 :")
    print("Employee1 is present only on : ",Employee4.count("P")," days")
    print("Employee1 is absent only on : ",Employee4.count("A")," days")
else:
  print("Employee present whole the week")
```
